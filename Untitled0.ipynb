```python
def get_completion(query: str, model, tokenizer) -> str:
  device = "cuda:0"

  prompt_template = """
  <start_of_turn>user
  Below is an instruction that describes a task. Write a response that appropriately completes the request.
  {query}
  <end_of_turn>\n<start_of_turn>model


  """
  prompt = prompt_template.format(query=query)

  encodeds = tokenizer(prompt, return_tensors="pt", add_special_tokens=True)

  model_inputs = encodeds.to(device)


  generated_ids = model.generate(**model_inputs, max_new_tokens=1000, do_sample=True, pad_token_id=tokenizer.eos_token_id)
  # decoded = tokenizer.batch_decode(generated_ids)
  decoded = tokenizer.decode(generated_ids[0], skip_special_tokens=True)
  return (decoded)
```

```python
def generate_prompt(data_point):
    prefix_text = 'As a professional academic reviewer, your task is to generate ratings that are consistent with human reviews based on the given paper information. Please carefully read the following paper information and generate ratings according to the specified format.\n\n'



    # Extract paper information
    prompt_lines = data_point['prompt'].split('\n')
    title = prompt_lines[0].split(': ')[1] if len(prompt_lines) > 0 and ': ' in prompt_lines[0] else ''
    abstract = prompt_lines[1].split(': ')[1] if len(prompt_lines) > 1 and ': ' in prompt_lines[1] else ''

    # Extract review comments
    completion_lines = data_point['completion'].split('\n')
    review = completion_lines[0].split(': ')[1] if len(completion_lines) > 0 and ': ' in completion_lines[0] else ''
    strengths_weaknesses = completion_lines[1].split(': ')[1] if len(completion_lines) > 1 and ': ' in completion_lines[1] else ''
    clarity_quality_novelty_reproducibility = completion_lines[2].split(': ')[1] if len(completion_lines) > 2 and ': ' in completion_lines[2] else ''
    correctness = completion_lines[3].split(': ')[1] if len(completion_lines) > 3 and ': ' in completion_lines[3] else ''
    technical_novelty_significance = completion_lines[4].split(': ')[1] if len(completion_lines) > 4 and ': ' in completion_lines[4] else ''
    empirical_novelty_significance = completion_lines[5].split(': ')[1] if len(completion_lines) > 5 and ': ' in completion_lines[5] else ''
    recommendation = completion_lines[6].split(': ')[1] if len(completion_lines) > 6 and ': ' in completion_lines[6] else ''

    # Adding Chain of Thought (COT) prompt
    cot_prompt = """Please think step-by-step and generate ratings:
        1. First, analyze the title and abstract of the paper to understand its main content and research direction.
        2. Then, read the review comments to understand the reviewer's focus and evaluation.
        3. Finally, based on the above information, generate ratings consistent with human reviews.

Paper information is as follows:

  Title: {}
  Abstract: {}

Review is as follows:

  Review: {}
    """.format(
        title,
        abstract,
        review,)



    text = f"""<start_of_turn>user {prefix_text} \n\n{cot_prompt} <end_of_turn>\n<start_of_turn>model{data_point["completion"]} <end_of_turn>"""
    return text



# add the "prompt" column in the dataset
text_column = [generate_prompt(data_point) for data_point in dataset]
print(text_column[0])
# dataset = dataset.add_column("prompt", text_column)
```

```python
import bitsandbytes as bnb
def find_all_linear_names(model):
  cls = torch.nn.Linear #if args.bits == 4 else (bnb.nn.Linear8bitLt if args.bits == 8 else torch.nn.Linear)
  lora_module_names = set()
  for name, module in model.named_modules():
    if isinstance(module, cls):
      names = name.split('.')
      lora_module_names.add(names[0] if len(names) == 1 else names[-1])
    if 'lm_head' in lora_module_names: # needed for 16-bit
      lora_module_names.remove('lm_head')
  return list(lora_module_names)
```

```python
modules = find_all_linear_names(model)
print(modules)
```

```python
from peft import LoraConfig, get_peft_model

lora_config = LoraConfig(
    r=64,
    lora_alpha=32,
    target_modules=modules,
    lora_dropout=0.05,
    bias="none",
    task_type="CAUSAL_LM"
)

model = get_peft_model(model, lora_config)
```

```python
#new code using SFTTrainer
import transformers

from trl import SFTTrainer

tokenizer.pad_token = tokenizer.eos_token
torch.cuda.empty_cache()

trainer = SFTTrainer(
    model=model,
    train_dataset=train_data,
    eval_dataset=test_data,
    dataset_text_field="prompt",
    peft_config=lora_config,
    args=transformers.TrainingArguments(
        per_device_train_batch_size=1,
        gradient_accumulation_steps=4,
        warmup_steps=0.03,
        max_steps=100,
        learning_rate=2e-4,
        logging_steps=1,
        output_dir="outputs",
        optim="paged_adamw_8bit",
        save_strategy="epoch",
    ),
    data_collator=transformers.DataCollatorForLanguageModeling(tokenizer, mlm=False),
)
```

```python
model.config.use_cache = False  # silence the warnings. Please re-enable for inference!
trainer.train()
```
